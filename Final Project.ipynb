### Drop kolom yang mempunyai lebih dari 50% missing value dari total data

```python
data.drop('company',axis=1,inplace=True)
```

### Mengisi kolom Agent dengan mean

```python
data['agent'] = data['agent'].fillna(data['agent'].mean())
```

### Mengisi kolom children dan country dengan nama yang sering muncul

```python
data['children'].value_counts()
```

```python
data['country'].value_counts()
```

```python
data['children'] = data['children'].fillna(0.0)
```

```python
data['children']
```

```python
data['children'].value_counts()
```

```python
data['country'] = data['country'].fillna('PRT')
```

```python
data['country']
```

```python
data.isna().sum()
```

### Membuat kategori berdasarkan waktu

```python
import numpy as np
```

```python
data['category'] = np.where(data['lead_time'] > 500, 'Long', np.where(data['lead_time'] > 200 , 'Medium', 'Short'))
```

```python
data.head()
```

## Filtering Data

###  Berdasarkan reservation_status_date > 2017

```python
data.dtypes
```

```python
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'],format = '%Y-%m-%d')
```

```python
start_date = "2017-1-1"
end_date = "2017-12-31"

after_start_date = data["reservation_status_date"] >= start_date
before_end_date = data["reservation_status_date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = data.loc[between_two_dates]
filtered_dates
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
month = ['January','February','March','April','May','June','July','August','September','Oktober','November','December']
visitor = [5929,8068,9794,11089,11791,10939,12661,13877,10508,11160,6794,6780]
```

```python
plt.figure(figsize=(13,6))
plt.plot(month, visitor,marker='o')
plt.xlabel('Bulan')
plt.ylabel('Pelanggan')
plt.title('Customer Visitor by Month')

plt.grid()
plt.show()
```

```python
month.reverse()
visitor.reverse()
plt.figure(figsize=(15,8))
sns.barplot(month,visitor)
plt.title('Customer Visitor by Month')
plt.xlabel('Bulan')
plt.savefig('plot2.png')
plt.show()
```
