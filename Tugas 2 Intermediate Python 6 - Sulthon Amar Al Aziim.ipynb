## Exercise 5

1. Create a new column with the following condition:\
    a. > 10 : Long Trip\
    b. 5-10 : Medium Trip\
    c. <5 : Short Trip

```python
import numpy as np
```

```python
#membuat list kondisi
kondisi = [(data['MILES*'] > 10),
          (data['MILES*'] >= 5) & (data['MILES*'] <= 10),
          (data['MILES*'] < 5)]

#membuat list nilai yang ingin dimasukkan untuk masing - masing kondisi
values = ['Long Trip','Medium Trip','Short Trip']

#membuat kolom baru dan menggunakan np.select untuk mendapatkan nilai yang sesuai kondisi
data['TYPE'] = np.select(kondisi,values)

data.head(10)
```
